```python
cmt_full = pd.DataFrame(df['comments_full'])
```

```python
cmt_full
```

```python
import datetime

def transform_comment(comment):
  return eval(comment)

cmt_full['comments_full'] = cmt_full['comments_full'].apply(transform_comment)
```

```python
cmt_full = cmt_full.explode("comments_full")
```

```python
comments_full = pd.json_normalize(cmt_full['comments_full'])
comments_full
```

```python
comments_full_cleaned = comments_full.dropna(axis=1, how='all')
comments_full_cleaned
```

```python
path = FOLDER_PATH + "comments.csv"
comments_full_cleaned.to_csv(path, index=False, escapechar='\\')
print(path)
```

```python
reactor = pd.DataFrame(df['reactors'])
```

```python
import datetime

def transform_comment(comment):
  return eval(comment)

reactor['reactors'] = reactor['reactors'].apply(transform_comment)
```

```python
reactor = reactor.explode("reactors")
```

```python
reactor_full = pd.json_normalize(reactor['reactors'])
reactor_full.dropna(inplace=True)
reactor_full
```

```python
FOLDER_PATH = "/content/drive/MyDrive/python"
path=FOLDER_PATH + "reactor.csv"
reactor_full.to_csv(path, index=False)
print(path)
```
